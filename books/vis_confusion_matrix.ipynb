```python
import numpy as np
import os

from sklearn import metrics 
import matplotlib.pyplot as plt
```

```python
#ck+ label
label = ['Neutral', 'Happiness', 'Surprise', 'Sadness', 'Anger', 'Disgust', 'Fear','Contempt']
```

```python
pred_file = "real_train_y.npz"
# pred_file = "real_val_y.npz"
# pred_file = "synthetic_train_y.npz"
# pred_file = "synthetic_val_y.npz"
```

```python
data = np.load(os.path.join(file_path, pred_file))
```

```python
y, yhat = data["name1"], data["name2"]
```

```python
cm = metrics.confusion_matrix(y, yhat)
cm_display = metrics.ConfusionMatrixDisplay(cm, display_labels=label).plot(
    xticks_rotation="vertical", values_format='d')
# plt.ticklabel_format(style='plain', axis='y',useOffset=False)
# plt.savefig(os.path.join(file_path, "ck_synthetic_cm.png"), dpi=800, bbox_inches='tight')
plt.savefig(os.path.join(file_path, "ck_real_cm.png"), dpi=800, bbox_inches='tight')
```
